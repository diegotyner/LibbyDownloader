```python
import json
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

options = Options()
options.set_capability('goog:loggingPrefs', {'performance': 'ALL'})
options.add_argument("--headless") 
driver = webdriver.Chrome(options=options)
driver.get("https://jsonplaceholder.typicode.com")
print(driver.title)
sleep(5)  # wait for the requests to take place

# extract requests from logs
logs_raw = driver.get_log("performance")
logs = [json.loads(lr["message"])["message"] for lr in logs_raw]

def log_filter(log_):
    return (
        # is an actual response
        log_["method"] == "Network.responseReceived"
        # and json
        and "json" in log_["params"]["response"]["mimeType"]
    )

for log in filter(log_filter, logs):
    request_id = log["params"]["requestId"]
    resp_url = log["params"]["response"]["url"]
    print(f"Caught {resp_url}")
    print(driver.execute_cdp_cmd("Network.getResponseBody", {"requestId": request_id}))

driver.quit()
```

```python
import json
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from time import sleep

options = Options()
options.add_argument("--headless")
options.set_capability('goog:loggingPrefs', {'performance': 'ALL'})

driver = webdriver.Chrome(options=options)
driver.get("https://jsonplaceholder.typicode.com/")

# Inject a known JSON API request
driver.execute_script("""
fetch("https://jsonplaceholder.typicode.com/posts/1")
""")
sleep(3)  # allow time for the request

logs_raw = driver.get_log("performance")
print("---- log start ---")
print(logs_raw)
print("---- logs done ---")
logs = [json.loads(entry["message"])["message"] for entry in logs_raw]

def is_json_response(log):
    return (
        log["method"] == "Network.responseReceived"
        and "json" in log["params"]["response"]["mimeType"]
    )

# Extract and print matching responses
for log in filter(is_json_response, logs):
    req_id = log["params"]["requestId"]
    url = log["params"]["response"]["url"]
    print(f"[+] JSON response from {url}")
    body = driver.execute_cdp_cmd("Network.getResponseBody", {"requestId": req_id})
    print(body)

driver.quit()
```

```python
import json

def try_json(value):
    try:
        return json.loads(value)
    except (json.JSONDecodeError, TypeError):
        return value

def deep_clean(d):
    return {k: try_json(v) for k, v in d.items()}


with open("libby_local_storage.txt", "r") as f:
  text = f.read()
  # print(text)
  no_quotes = text.replace("\\\\", "\\")
  # print(no_quotes)

  parsed = json.loads(no_quotes)
  cleaned = deep_clean(parsed)
  # print(json.dumps(cleaned, indent=2))

  with open('libby_local_storage.json', 'w') as f:
    json.dump(cleaned, f)

```
